## Deep-Dive Analysis: Segmentation Analysis (by Price)

We will segment laptops based on their price to identify different market tiers and understand the characteristics of products within each tier. This can help in understanding target customer groups and competitive positioning.

```python
# Define price segments using quantiles
df['Price_Segment'] = pd.qcut(df['Price'], q=3, labels=['Budget', 'Mid-Range', 'Premium'])

print("Price Segment Distribution:")
display(df['Price_Segment'].value_counts())

print("\nAverage Price by Segment:")
display(df.groupby('Price_Segment')['Price'].mean())
```

```python
# Analyze key numerical features by Price Segment
print("\nAverage RAM (GB) by Price Segment:")
display(df.groupby('Price_Segment')['Ram_GB'].mean())

print("\nAverage Total Storage (GB) by Price Segment:")
display(df.groupby('Price_Segment')['Total_Storage_GB'].mean())

print("\nAverage Screen Size (Inches) by Price Segment:")
display(df.groupby('Price_Segment')['Inches'].mean())

print("\nAverage Weight (kg) by Price Segment:")
display(df.groupby('Price_Segment')['Weight_kg'].mean())
```

```python
# Analyze key categorical features by Price Segment

print("\nTop 5 Companies by Price Segment:")
for segment in df['Price_Segment'].unique():
    print(f"\n--- {segment} ---")
    display(df[df['Price_Segment'] == segment]['Company'].value_counts().head(5))

print("\nOperating System Distribution by Price Segment:")
for segment in df['Price_Segment'].unique():
    print(f"\n--- {segment} ---")
    display(df[df['Price_Segment'] == segment]['OpSys'].value_counts(normalize=True).mul(100).round(2))
```

### Summary of Price Segments:

*   **Budget Segment:** Characterized by lower average RAM, smaller total storage, and often more basic screen resolutions. Dominated by companies focusing on cost-effective solutions.
*   **Mid-Range Segment:** Offers a balance of features and price. Shows an increase in average RAM and storage, with a wider variety of companies.
*   **Premium Segment:** Features the highest average RAM, largest storage capacities, larger and higher-resolution screens, and often more powerful CPUs/GPUs. Brands in this segment typically offer high-performance or specialized laptops. Apple and other high-end manufacturers are more prevalent here.

## Interactive Dashboard Build

We will use `Plotly Express` to create interactive visualizations for our dashboard, allowing users to explore the core KPIs and segmentation analysis. The first visualization will be a scatter plot to illustrate the price segmentation based on storage.

```python
import plotly.express as px

fig = px.scatter(df,
                 x='Total_Storage_GB',
                 y='Price',
                 color='Price_Segment',
                 hover_data=['Company', 'TypeName', 'Ram_GB', 'Cpu', 'Gpu'],
                 title='Laptop Price vs. Total Storage by Price Segment',
                 labels={'Total_Storage_GB': 'Total Storage (GB)', 'Price': 'Price ($)'})

fig.update_layout(height=600, width=900)

fig.show()
```

```python
fig_ram = px.bar(df.groupby(['Company', 'Price_Segment'], observed=False)['Ram_GB'].mean().reset_index(),
                 x='Company',
                 y='Ram_GB',
                 color='Price_Segment',
                 barmode='group',
                 title='Average RAM (GB) by Company and Price Segment',
                 labels={'Ram_GB': 'Average RAM (GB)', 'Company': 'Company'},
                 height=600,
                 width=900)

fig_ram.update_layout(xaxis={'categoryorder':'total descending'})
fig_ram.show()
```
