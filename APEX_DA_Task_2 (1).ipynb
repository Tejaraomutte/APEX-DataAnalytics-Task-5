## Exploratory Data Analysis (EDA) for Gaming & Academic Performance

### Step 1: Load Data and Initial Inspection

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/Gaming_Academic_Performance.csv')

print("Dataset loaded successfully. Displaying the first 5 rows:")
display(df.head())
```

Let's check the general information about the dataset, including data types and non-null values.

```python
print("\nDataset Information:")
df.info()
```

### Step 2: Descriptive Statistics & Univariate Analysis

First, let's calculate key summary statistics for numerical fields.

```python
print("\nDescriptive Statistics for Numerical Columns:")
display(df.describe())
```

Next, we'll create initial visualizations (histograms for numerical data and bar charts for categorical data) to understand the distributions of individual variables.

```python
# Identify numerical and categorical columns
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

# Create histograms for numerical columns
print("\nHistograms for Numerical Columns:")
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Create bar charts for categorical columns
print("\nBar Charts for Categorical Columns:")
for col in categorical_cols:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index)
    plt.title(f'Count of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()
```

### Step 3: Multivariate Analysis & Correlation

Now, let's explore relationships between two or more variables by creating advanced visualizations like scatter plots and heatmaps.

```python
print("\nCorrelation Matrix of Numerical Columns:")
correlation_matrix = df[numerical_cols].corr()
display(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

print("\nScatter Plots for key relationships (e.g., Gaming vs Academic Performance):")
# Example scatter plots for potentially interesting relationships

# Assuming 'HoursPlayed' and 'GPA' are present based on typical academic performance datasets
if 'HoursPlayed' in df.columns and 'GPA' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x='HoursPlayed', y='GPA', hue='Gender')
    plt.title('Hours Played vs. GPA')
    plt.xlabel('Hours Played (per week/day)')
    plt.ylabel('GPA')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# You can add more scatter plots based on the specific columns available in your dataset
# For example, if there are other relevant numerical columns, you can add more plots here.

# Pair plot can be very informative but might be heavy for many columns.
# Let's consider a subset if the number of numerical columns is large.
if len(numerical_cols) <= 7 and len(numerical_cols) > 1:
    print("\nPair Plot for Numerical Columns:")
    sns.pairplot(df[numerical_cols].dropna())
    plt.suptitle('Pair Plot of Numerical Features', y=1.02)
    plt.show()
elif len(numerical_cols) > 7:
    print("\nPair plot skipped due to a large number of numerical columns. Consider selecting a subset for pair plots.")
```

### Business Questions for SQL Query Practice

Here are some business questions for the intern to practice SQL queries, focusing on filtering, aggregation, and relationships within the `Gaming_Academic_Performance.csv` dataset:

1.  **Academic Performance by Gender:** What is the average `grades` for male and female students in the dataset?
2.  **Gaming Hours vs. Grades:** How does the average `grades` vary for students grouped by different ranges of `gaming_hours` (e.g., 0-5 hours, 5-10 hours, 10-15 hours, etc.)?
3.  **Addiction Score vs. Performance:** What are the average `grades` and `study_hours` for students categorized into 'low', 'medium', and 'high' `addiction_score` groups? (The intern may need to define these thresholds based on the data distribution).
4.  **Gaming Genre Impact:** Which `gaming_genre` is associated with the highest and lowest average `grades` among students?
5.  **Sleep, Gaming, and Grades Interplay:** For students who engage in high `gaming_hours` (e.g., more than 10 hours per day), what is the relationship between their `sleep_hours` and their `grades`? (This requires filtering and then aggregation).
6.  **Stress Level and Academic Outcomes:** What is the average `grades` for students at each `stress_level` category (`Low`, `Medium`, `High`)?
7.  **Device Usage and Attendance/Grades:** How do average `grades` and `attendance` differ across different quartiles or ranges of `device_usage`?

### Answering Business Questions using SQL (via `pandasql`)

```python
import pandasql as ps

# Question 1: Academic Performance by Gender
query1 = """
SELECT gender, AVG(grades) AS average_grades
FROM df
GROUP BY gender
"""
result1 = ps.sqldf(query1, globals())
print("1. Academic Performance by Gender:")
display(result1)
```

```python
# Question 2: Gaming Hours vs. Grades
# Define gaming hour ranges
query2 = """
SELECT
  CASE
    WHEN gaming_hours BETWEEN 0 AND 5 THEN '0-5 hours'
    WHEN gaming_hours > 5 AND gaming_hours <= 10 THEN '5-10 hours'
    WHEN gaming_hours > 10 AND gaming_hours <= 15 THEN '10-15 hours'
    WHEN gaming_hours > 15 AND gaming_hours <= 20 THEN '15-20 hours'
    ELSE '>20 hours'
  END AS gaming_hours_range,
  AVG(grades) AS average_grades
FROM df
GROUP BY gaming_hours_range
ORDER BY MIN(gaming_hours)
"""
result2 = ps.sqldf(query2, globals())
print("\n2. Gaming Hours vs. Grades:")
display(result2)
```

```python
# Question 3: Addiction Score vs. Performance
# Assuming addiction_score ranges from 0-100 based on typical scoring. Adjust thresholds as needed.
query3 = """
SELECT
  CASE
    WHEN addiction_score < 33.33 THEN 'Low Addiction'
    WHEN addiction_score >= 33.33 AND addiction_score < 66.67 THEN 'Medium Addiction'
    ELSE 'High Addiction'
  END AS addiction_level,
  AVG(grades) AS average_grades,
  AVG(study_hours) AS average_study_hours
FROM df
GROUP BY addiction_level
ORDER BY AVG(addiction_score)
"""
result3 = ps.sqldf(query3, globals())
print("\n3. Addiction Score vs. Performance:")
display(result3)
```

```python
# Question 4: Gaming Genre Impact
query4 = """
SELECT gaming_genre, AVG(grades) AS average_grades
FROM df
GROUP BY gaming_genre
ORDER BY average_grades DESC
"""
result4 = ps.sqldf(query4, globals())
print("\n4. Gaming Genre Impact:")
display(result4)
```

```python
# Question 5: Sleep, Gaming, and Grades Interplay (for high gaming hours)
# Define 'high gaming_hours' as > 10 hours for this example.
query5 = """
SELECT
  CASE
    WHEN sleep_hours <= 6 THEN 'Low Sleep'
    WHEN sleep_hours > 6 AND sleep_hours <= 8 THEN 'Medium Sleep'
    ELSE 'High Sleep'
  END AS sleep_category,
  AVG(grades) AS average_grades
FROM df
WHERE gaming_hours > 10
GROUP BY sleep_category
ORDER BY MIN(sleep_hours)
"""
result5 = ps.sqldf(query5, globals())
print("\n5. Sleep, Gaming, and Grades Interplay (for high gaming hours > 10):")
display(result5)
```

```python
# Question 6: Stress Level and Academic Outcomes
query6 = """
SELECT stress_level, AVG(grades) AS average_grades
FROM df
GROUP BY stress_level
ORDER BY average_grades DESC
"""
result6 = ps.sqldf(query6, globals())
print("\n6. Stress Level and Academic Outcomes:")
display(result6)
```

```python
# Question 7: Device Usage and Attendance/Grades
# Let's use quartiles for device_usage. First, calculate quartiles in Python then use in SQL.
device_usage_quartiles = df['device_usage'].quantile([0.25, 0.5, 0.75]).tolist()

query7 = f"""
SELECT
  CASE
    WHEN device_usage <= {device_usage_quartiles[0]} THEN 'Q1'
    WHEN device_usage > {device_usage_quartiles[0]} AND device_usage <= {device_usage_quartiles[1]} THEN 'Q2'
    WHEN device_usage > {device_usage_quartiles[1]} AND device_usage <= {device_usage_quartiles[2]} THEN 'Q3'
    ELSE 'Q4'
  END AS device_usage_quartile,
  AVG(grades) AS average_grades,
  AVG(attendance) AS average_attendance
FROM df
GROUP BY device_usage_quartile
ORDER BY MIN(device_usage)
"""
result7 = ps.sqldf(query7, globals())
print("\n7. Device Usage and Attendance/Grades:")
display(result7)
```
